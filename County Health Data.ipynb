#**Looking at Attributes and Indexing Our Dataset**

Now that our data has been uploaded, named, and made readable to Pandas, we will need to narrow our range to begin indexing and making use of specific data.

But first, it would be helpful to look at the attributes of our dataset.

To begin, we will use the `.shape` and `.columns` functions to help us  narrow our search.

* type `CHD.shape` into an empty cell, and press **SHIFT** + **ENTER** to execute it.

```python
CHD.shape
```

Repeat with `CHD.columns`.

```python
CHD.columns
```

From these functions, we have a better sense of the size and what kind of data we will be working with.

For example, we know using the `.shape` function the number of of rows and colums are contained in the dataset **_(6109, 64)_**.

We also know using the `.columns` function what some of the categories we can explore look like.

>For our purposes here, we will be looking data in the **_"Region"_**, **_"Uninsured"_**, **_"Uninsured adults"_**, **_"Income inequality"_**, and **_"Preventable hospital stays"_** columns.

Next, we will find the specific columns listed above using the `.loc` function.

This will provide a full list of those specific data points.

* we will type
`CHD.loc[:,["Region","Uninsured","Uninsured adults","Income inequality","Preventable hospital stays"]]` in a cell.

Make sure this is typed precisely to avoid receiving an error message.

Press **SHIFT** + **ENTER** to execute.

```python
CHD.loc[:,["Region","Uninsured","Uninsured adults","Income inequality","Preventable hospital stays"]]
```

Now that we have culled this subset, to make future analysis easier, it would be helpful to rename it.

Instead of having to to type `"CHD.loc[:,"Region","Uninsured","Uninsured adults","Income inequality","Preventable hospital stays"]]` each time we want to look at these specific columns, we can shorten it.

For our purposes, we'll rename it '**_RUUI_subset_**'.

To do this, we will type `RUUI_subset = CHD.loc[:,["Region","Uninsured","Uninsured adults","Income inequality","Preventable hospital stays"]]`

```python
RUUI_subset = CHD.loc[:,["Region","Uninsured","Uninsured adults","Income inequality","Preventable hospital stays"]]
```

To be certain, we can simply enter our new subset into a cell and execute it.

It should provide the same table.

```python
RUUI_subset
```

Now that we have our data subset, we can narrow our focus.

Perhaps we want to look at the regional data for the Southern United States.

* to complete this command, we will type
`RUUI_subset["Region"] == "South"]`

This should produce data for only the South.

```python
RUUI_subset[RUUI_subset["Region"] == "South"]
```

The same search can be done only for the Northeast as well!

Simply replace '**_South_**' with '**_Northeast_**' using the same function.

```python
RUUI_subset[RUUI_subset["Region"] == "Northeast"]
```

> We will notice uneven distribution of data from the Southern United States and the Northeast, which may affect our analysis.

We may want to narrow our search further and look at random sample data from a specific region.

To do this, we will use our data subset (RUUI_subset), and the `.sample` function.

To execute this, we will type `RUUI_subset[RUUI_subset["Region"] == "South"].sample(n=20)`.

This will provide us random sample data of twenty observations from the Southern United States.

```python
RUUI_subset[RUUI_subset["Region"] == "South"].sample(n=20)
```

We can can execute the same function for the Northeastern United States using the same function.

* we will type `RUUI_subset[RUUI_subset["Region"] == "Northeast"].sample(n=20)`

This will provide us with similar random sample data.

```python
RUUI_subset[RUUI_subset["Region"] == "Northeast"].sample(n=20)
```

We may also want to look at the median numbers from each region.

* to execute this, we will type `RUUI_subset[RUUI_subset["Region"] == "South"].median()`

This will give us the median numbers from the entire Southern region.

```python
RUUI_subset[RUUI_subset["Region"] == "South"].median()
```

```python
RUUI_subset[RUUI_subset["Region"] == "Northeast"].median()
```
